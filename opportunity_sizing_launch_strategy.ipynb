```python
competitors = pd.DataFrame({
    "Competitor": ["Bajaj Finance","LazyPay","Simpl","Amazon Pay Later","Flipkart Pay Later","Slice"],
    "Users_M": [88,15,10,20,18,14],
    "Merchants": [190000,50000,30000,80000,60000,25000]
})

print(competitors)
```

```python
plt.figure(figsize=(10,6))

colors = ["red" if x=="Bajaj Finance" else "steelblue"
          for x in competitors["Competitor"]]

plt.scatter(
    competitors["Merchants"],
    competitors["Users_M"],
    s=competitors["Users_M"]*8,
    c=colors,
    alpha=0.7
)

for _, row in competitors.iterrows():

    plt.annotate(
        row["Competitor"],
        (row["Merchants"], row["Users_M"]),
        xytext=(5,5),
        textcoords="offset points"
    )

plt.xlabel("Merchant Network")
plt.ylabel("User Base (Millions)")
plt.title("BNPL Competitive Positioning Matrix")

plt.grid(alpha=0.3)

plt.show()
```

KEY INSIGHTS
1. Bajaj Finance is market leader
2. Market concentrated around few players
3. Merchant network is key moat
4. Opportunity exists in Tier-2 BNPL segment
5. New entrant should target underserved users
