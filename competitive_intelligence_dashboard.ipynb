```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')

print("Libraries Loaded")
```

```python
STATE_DATA = pd.read_csv("../DATA/4_State_Data.csv")

CUSTOMER_SEGMENTS = pd.read_csv("../DATA/7_Customer_Segments.csv")

MERCHANT_SEGMENTS = pd.read_csv("../DATA/9_Merchant_Segments.csv")

COMPETITORS = pd.read_csv("../DATA/5_Competitors.csv")

UNIT_ECONOMICS = pd.read_csv("../DATA/10_Unit_Economics.csv")

BNPL_MARKET = pd.read_csv("../DATA/6_BNPL_Market.csv")
```

```python
TOP_STATES = STATE_DATA.sort_values(
    by="Population_M",
    ascending=False
)

TOP_STATES.head(10)
```

```python
TOP5 = TOP_STATES.head(5)

plt.figure(figsize=(10,5))

plt.bar(
    TOP5["State"],
    TOP5["Population_M"]
)

plt.title("Top 5 Launch Markets")

plt.xlabel("State")

plt.ylabel("Population (M)")

plt.show()
```

```python
STATE_DATA["OPPORTUNITY_SCORE"] = (
    STATE_DATA["Population_M"] /
    STATE_DATA["Population_M"].mean()
).round(2)

STATE_DATA["PRIORITY"] = np.where(
    STATE_DATA["OPPORTUNITY_SCORE"] >= 1.2,
    "HIGH",
    "MEDIUM"
)

STATE_DATA[
    ["State","Population_M",
     "OPPORTUNITY_SCORE","PRIORITY"]
].sort_values(
    "OPPORTUNITY_SCORE",
    ascending=False
)
```

```python
plt.figure(figsize=(10,5))

plt.bar(
    STATE_DATA["State"],
    STATE_DATA["OPPORTUNITY_SCORE"]
)

plt.title("State Opportunity Score")
plt.xticks(rotation=45)

plt.show()
```

Key Finding:

Uttar Pradesh, Maharashtra and Bihar
have the highest opportunity scores.

These states should be prioritised
for BNPL expansion.
