```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")

plt.style.use('seaborn-v0_8-whitegrid')

plt.rcParams['figure.figsize']=(10,6)
plt.rcParams['font.size']=11

print("Libraries Loaded Successfully")
```

```python
import pandas as pd

STATE_DATA = pd.read_csv("../DATA/4_State_Data.csv")
COMPETITORS = pd.read_csv("../DATA/5_Competitors.csv")
BNPL_MARKET = pd.read_csv("../DATA/6_BNPL_Market.csv")
UPI_DATA = pd.read_csv("../DATA/2_UPI_Data.csv")

print("STATE DATA :", STATE_DATA.shape)
print("COMPETITORS :", COMPETITORS.shape)
print("BNPL MARKET :", BNPL_MARKET.shape)
print("UPI DATA :", UPI_DATA.shape)
```

```python
print("STATE DATA")
print(STATE_DATA.columns)

print("\nCOMPETITORS")
print(COMPETITORS.columns)

print("\nBNPL MARKET")
print(BNPL_MARKET.columns)

print("\nUPI DATA")
print(UPI_DATA.columns)
```

```python
print("="*50)
print("INDIA BNPL MARKET ATTRACTIVENESS")
print("="*50)

print(BNPL_MARKET)
```

```python
TOP_STATES = (
    STATE_DATA
    .sort_values("Population_M", ascending=False)
)

TOP_STATES.head(10)
```

```python
TOP10 = TOP_STATES.head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    data=TOP10,
    x="State",
    y="Population_M"
)

plt.title("Top 10 States by Population")
plt.xlabel("State")
plt.ylabel("Population (Millions)")
plt.xticks(rotation=45)

plt.show()
```

```python
COMPETITOR_RANKING = (
    COMPETITORS
    .sort_values("Users_M", ascending=False)
)

COMPETITOR_RANKING
```

```python
plt.figure(figsize=(10,5))

sns.barplot(
    data=COMPETITOR_RANKING,
    x="Competitor",
    y="Users_M"
)

plt.title("BNPL Competitor User Base")
plt.xlabel("Competitor")
plt.ylabel("Users (Millions)")
plt.xticks(rotation=45)

plt.show()
```

```python
MERCHANT_RANKING = (
    COMPETITORS
    .sort_values("Merchants", ascending=False)
)

MERCHANT_RANKING
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    data=MERCHANT_RANKING,
    x="Competitor",
    y="Merchants"
)

plt.title("Merchant Network Comparison")
plt.xlabel("Competitor")
plt.ylabel("Merchant Count")
plt.xticks(rotation=45)

plt.show()
```

```python
AVG_POP = STATE_DATA["Population_M"].mean()

STATE_DATA["OPPORTUNITY_INDEX"] = round(
    STATE_DATA["Population_M"] / AVG_POP,
    2
)

STATE_DATA.head()
```

```python
OPPORTUNITY = (
    STATE_DATA
    .sort_values(
        "OPPORTUNITY_INDEX",
        ascending=False
    )
    .head(10)
)

plt.figure(figsize=(12,6))

sns.barplot(
    data=OPPORTUNITY,
    x="State",
    y="OPPORTUNITY_INDEX"
)

plt.title("State Opportunity Index")
plt.xlabel("State")
plt.ylabel("Opportunity Index")

plt.xticks(rotation=45)

plt.show()
```

```python
UPI_PERCENT = UPI_DATA.iloc[1:].copy()

UPI_PERCENT["Value"] = (
    UPI_PERCENT["Value"]
    .astype(str)
    .str.replace("%","", regex=False)
    .astype(float)
)

print(UPI_PERCENT.dtypes)
```

```python
plt.figure(figsize=(10,5))

sns.barplot(
    data=UPI_PERCENT,
    x="Metric",
    y="Value"
)

plt.title("UPI Ecosystem Readiness (%)")
plt.ylabel("Percentage")
plt.xlabel("")
plt.xticks(rotation=30)

plt.tight_layout()
plt.show()
```
